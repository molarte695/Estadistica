```python
import matplotlib.pyplot as plt
import numpy as np

# Get the data for the pie chart
pie_counts = df_filtrado['Area_Nacimiento'].value_counts().sort_index()
labels = pie_counts.index
sizes = pie_counts.values

fig, ax = plt.subplots(figsize=(10, 9)) # Increased figure width for legend

# Plot the pie chart without automatic percentage labels
wedges, texts = ax.pie(
    sizes,
    autopct=None, # Disable automatic percentage labels
    startangle=90,
    colors=plt.cm.tab20.colors,
    radius=0.8 # Reduced radius to make the pie smaller
)

# Adjust title font size and padding
ax.set_title('Distribución por Área de Nacimiento (Orden Alfabético)', pad=20, fontsize=14)

# Add legend with smaller font size and positioned outside
ax.legend(wedges, labels,
          title="Área de Nacimiento",
          loc="center left",
          bbox_to_anchor=(1.05, 0.5),
          fontsize=8,
          title_fontsize=10)

# Manually add percentage labels and arrows
for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1 # Mid-angle of the wedge
    y_coord = np.sin(np.deg2rad(ang))
    x_coord = np.cos(np.deg2rad(ang))

    percent_value = (sizes[i] / sum(sizes)) * 100

    # Determine position for the percentage text and arrow end point
    if percent_value < 2: # Aggressively push out labels for small slices
        text_radius_multiplier = 2.2 # Further increased from 1.9
    else:
        text_radius_multiplier = 1.8 # Further increased from 1.5

    # Arrow start point (on the edge of the pie slice)
    arrow_start_x = p.r * x_coord
    arrow_start_y = p.r * y_coord

    # Text end point (where the percentage text will be placed)
    text_end_x = p.r * text_radius_multiplier * x_coord
    text_end_y = p.r * text_radius_multiplier * y_coord

    # Determine text alignment
    horizontalalignment = 'center'
    if x_coord > 0.05: # right side
        horizontalalignment = 'left'
    elif x_coord < -0.05: # left side
        horizontalalignment = 'right'

    ax.annotate(f"{percent_value:.1f}%",
                xy=(arrow_start_x, arrow_start_y), # Point on the pie circumference
                xytext=(text_end_x, text_end_y),   # Position of the text
                ha=horizontalalignment,
                va='center',
                fontsize=10,
                weight='bold',
                arrowprops=dict(arrowstyle="-", color='gray', linewidth=0.7))

ax.set_ylabel('')

# Manually adjust subplot parameters to make room for the legend
fig.subplots_adjust(right=0.7)

plt.show()
```

```python
# Convertir la columna 'Peso_Grs.' a tipo numérico, reemplazando la coma por un punto
df_filtrado['Peso_Grs.'] = df_filtrado['Peso_Grs.'].str.replace(',', '.').astype(float)

# Calcular Q1, Q3 e IQR para ambas variables en un solo paso
Q1 = df_filtrado[['Talla_cms', 'Peso_Grs.']].quantile(0.25)
Q3 = df_filtrado[['Talla_cms', 'Peso_Grs.']].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para considerar un dato como atípico
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar el DataFrame para eliminar outliers en ambas variables
condicion = (
    (df_filtrado['Talla_cms'] >= limite_inferior['Talla_cms']) & (df_filtrado['Talla_cms'] <= limite_superior['Talla_cms']) &
    (df_filtrado['Peso_Grs.'] >= limite_inferior['Peso_Grs.']) & (df_filtrado['Peso_Grs.'] <= limite_superior['Peso_Grs.'])
)
df_sinatipicos = df_filtrado[condicion]

# Mostrar la cantidad de datos restantes
print(f"Número de datos restantes después de eliminar outliers: {len(df_sinatipicos)}")
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest, anderson, jarque_bera, probplot

# Lista de variables a evaluar (corregidas a los nombres del DataFrame)
columnas_fisicas = ['Talla_cms', 'Peso_Grs.']

# Diccionario para almacenar resultados
resultados_normalidad = {}

# Evaluar cada variable
for col in columnas_fisicas:
    # Usar df_sinatipicos para los tests de normalidad
    if col in df_sinatipicos.columns:
        datos = df_sinatipicos[col].dropna()
        # Shapiro-Wilk
        stat_sw, p_sw = shapiro(datos)
        # Kolmogorov-Smirnov
        media, desv = np.mean(datos), np.std(datos)
        stat_ks, p_ks = kstest(datos, "norm", args=(media, desv))
        # Anderson-Darling
        resultado_ad = anderson(datos)
        stat_ad = resultado_ad.statistic
        crit_ad = resultado_ad.critical_values[2]  # 5%
        ad_normal = stat_ad < crit_ad
        # Jarque-Bera
        stat_jb, p_jb = jarque_bera(datos)
        # Guardar resultados
        resultados_normalidad[col] = {
            'Shapiro-Wilk': p_sw,
            'Kolmogorov-Smirnov': p_ks,
            'Anderson-Darling': f'Estadístico: {stat_ad:.4f}, Crítico (5%): {crit_ad:.4f}',
            'Jarque-Bera': p_jb,
            'Normal': sum([
                p_sw > 0.05,
                p_ks > 0.05,
                ad_normal,
                p_jb > 0.05
            ]) >= 3
        }

# Mostrar resultados en tabla
print("Resultados de los tests de normalidad (valores p):\n")
for var, res in resultados_normalidad.items():
    print(f"Variable: {var}")
    print(f"  Shapiro-Wilk p = {res['Shapiro-Wilk']:.4f}")
    print(f"  Kolmogorov-Smirnov p = {res['Kolmogorov-Smirnov']:.4f}")
    print(f"  Anderson-Darling {res['Anderson-Darling']}")
    print(f"  Jarque-Bera p = {res['Jarque-Bera']:.4f}")
    if res['Normal']:
        print("  → Distribución normal (según mayoría de tests)\n")
    else:
        print("  → NO es distribución normal (según mayoría de tests)\n")

# Gráficos QQ para cada variable
fig, axes = plt.subplots(1, len(columnas_fisicas), figsize=(7 * len(columnas_fisicas), 6))

if len(columnas_fisicas) == 1:
    axes = [axes]

for i, col in enumerate(columnas_fisicas):
    if col in df_sinatipicos.columns:
        datos = df_sinatipicos[col].dropna()
        probplot(datos, dist="norm", plot=axes[i])
        axes[i].set_title(f'QQ Plot de {col}')

plt.tight_layout()
plt.show()
```
